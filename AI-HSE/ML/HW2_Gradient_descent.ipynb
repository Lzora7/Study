```python
def eval_func(func, x_values):
  '''
  значение функции
  '''
  y_values = []
  for value in x_values:
    y_value = func.subs(x, value).evalf()
    y_values.append(y_value)

  return y_values
```

```python
def eval_f_a(x):
    return x**2 - 30*x + 1

def eval_f_b(x):
    return -x**3 + 5*x - 7*x**2 + 17

def eval_f_c(x):
    return x**2 - 1/x

def eval_f_d(x):
    return 1/x**5 + x**17 + 8
```

```python
def diff_1(func):
  '''
  взятие I производной
  '''
  diffed = sp.diff(func)
  return diffed
```

```python
def diff_2(func):
  '''
  взятие II производной
  '''
  diffed = sp.diff(func, x, 2)
  return diffed
```

```python
def extremum_x(func):
  '''
  значения x в экстремумах
  '''
  diff_func = diff_1(func)
  extremum_points = sp.solve(diff_func, x)
  return extremum_points
```

```python
def peregib(f):
  # вторая пр
  f_double_prime = diff_2(f)

  # нули второй производной
  inflection_points = sp.solve(f_double_prime, x)

  points_x = []
  # подставим в точки слева и справа
  for point in inflection_points:

    test_value = point - 0.01
    left_sign = f_double_prime.subs(x, test_value).evalf()

    test_value = point + 0.01
    right_sign = f_double_prime.subs(x, test_value).evalf()

    if left_sign * right_sign < 0:
        points_x.append(point)

  return points_x
```

Модифицируйте функцию, осуществляющую метод градиентного спуска с семинара: добавьте опцию останавливать метод, если абсолютное значение производной функции в точке < некоторого $\delta$ (delta добавьте в аргументы функции).

```python
# your code here
```

```python
def f(x):
  return x**3 - 3*x**2 + 4
```

```python
def gradient_descent(x_start, learning_rate, epsilon, delta, num_iterations):
    x_curr = x_start
    df_x = sp.diff(f(x)) # производная

    trace = []
    trace.append(x_curr)

    for i in range(num_iterations):
      x_new = x_curr - learning_rate * df_x.subs(x, x_curr)
      trace.append(x_new)

      if abs(x_new - x_curr) < epsilon:
        return x_curr, trace

      # added
      if abs(df_x.subs(x, x_curr) - df_x.subs(x, x_new)) < delta:
        return x_curr, trace

      x_curr = x_new

    return x_curr, trace
```

```python
# delta = 1
xmin, trace = gradient_descent(3, 0.01, 0.001, 1, 10)

x_values = [x for x in np.arange(-5, 5, 0.1)]
f_values = [f(x) for x in x_values]

plt.figure(figsize=(10,10))

plt.axvline(x=0, c = 'black')
plt.axhline(y=0, c = 'black')

plt.plot(x_values, f_values)

plt.xlim([-3, 5])
plt.ylim([-5, 7])

plt.title('График функции f(x) c экстремумами и точками перегиба')

plt.xlabel('x')
plt.ylabel('f(x)')

trace_values = [f(x) for x in trace]
plt.scatter(trace, trace_values, c='red')

plt.show()
```
